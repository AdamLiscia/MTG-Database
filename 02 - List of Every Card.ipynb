```python
# YouTube Video: https://www.youtube.com/watch?v=Z3vFdtZ7d-g
from selenium import webdriver
import pandas as pd
import time

start = time.time()

MAX_PAGE_NUM = 196

driver = webdriver.Chrome(executable_path='C:/Users/adaml/Downloads/chromedriver_win32/chromedriver.exe')

names = []

for i in range(0, MAX_PAGE_NUM + 1):

    url = "http://gatherer.wizards.com/Pages/Search/Default.aspx?page=" + str(i) + "&output=standard&action=advanced&cmc=+%3E=[0]"
    driver.get(url)

    names_scrape = driver.find_elements_by_xpath("//span[@class='cardTitle']")
    names_lst = [x.text for x in names_scrape]
    names.extend(names_lst)
    

magic_df = pd.DataFrame(
        {'Name': names})


driver.close()

end = time.time()
print (end - start)
```
